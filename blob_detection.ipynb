```python
#mask
import cv2
import numpy as np

# Initialize the webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Capture a frame from the webcam
    ret, frame = video_capture.read()

    if not ret:
        break

    # Convert the frame from BGR to HSV color space
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the range of blue color in HSV
    lower_blue = np.array([100, 100, 50])
    upper_blue = np.array([130, 255, 255])

    # Create a binary mask for blue regions
    mask = cv2.inRange(hsv_frame, lower_blue, upper_blue)

    # Apply morphological operations to clean up the mask
    kernel = np.ones((5, 5), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    # Overlay the blue regions on the original frame
    blue_regions = cv2.bitwise_and(frame, frame, mask=mask)
    
    
    # Display the resulting frame
    cv2.imshow('Object Detection', blue_regions)
    # Display the original frame and the blue regions
    #cv2.imshow('Original', frame)
    #cv2.imshow('Blue Regions', result)
    

    # Exit the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
video_capture.release()
cv2.destroyAllWindows()
```

```python
#contour detection
import cv2
import numpy as np

# Initialize the webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Capture a frame from the webcam
    ret, frame = video_capture.read()

    if not ret:
        break

    # Convert the frame from BGR to HSV color space
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the range of blue color in HSV
    lower_blue = np.array([100, 100, 50])
    upper_blue = np.array([130, 255, 255])

    # Create a binary mask for blue regions
    mask = cv2.inRange(hsv_frame, lower_blue, upper_blue)

    # Find contours of the blue regions in the mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw boundary boxes around the blue regions
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Blue Color Segmentation', frame)

    # Exit the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
video_capture.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
# Initialize the webcam
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert the frame to the HSV color space
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the lower and upper bounds of the blue color in HSV
    lower_blue = np.array([90, 50, 50])  # Adjust these values to define the blue color range
    upper_blue = np.array([130, 255, 255])

    # Create a binary mask for the blue color range
    blue_mask = cv2.inRange(hsv_frame, lower_blue, upper_blue)

    # Apply the binary mask to extract the blue region from the frame
    blue_frame = cv2.bitwise_and(frame, frame, mask=blue_mask)

    # Convert the blue frame to grayscale
    gray_frame = cv2.cvtColor(blue_frame, cv2.COLOR_BGR2GRAY)

    # Apply the LoG operator to detect blobs
    ksize = 1  # Adjust the kernel size based on the size of blobs you want to detect
    sigma = 2.5  # Adjust the sigma value based on the spread of blob intensity
    blobs = cv2.Laplacian(gray_frame, cv2.CV_64F, ksize=ksize)
    blobs = np.uint8(np.absolute(blobs))

    # Threshold the blob image to filter out low-intensity regions
    threshold = 200 # Adjust the threshold value as per your requirements
    _, binary_blobs = cv2.threshold(blobs, threshold, 255, cv2.THRESH_BINARY)

    # Display the original frame and blob-detected frame
    cv2.imshow('Original Frame', frame)
    cv2.imshow('Blue Blob Detected Frame', binary_blobs)

    # Exit the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

def blob_detection_blue_background(frame):
    # Convert the frame to the HSV color space
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the lower and upper bounds of the blue color range
    lower_blue = np.array([100, 50, 50])  # Adjust these values for your specific shade of blue
    upper_blue = np.array([130, 255, 255])  # Adjust these values for your specific shade of blue

    # Create a mask to extract only blue regions from the frame
    blue_mask = cv2.inRange(hsv_frame, lower_blue, upper_blue)

    # Apply the mask to the original frame to get the blue background
    blue_background = cv2.bitwise_and(frame, frame, mask=blue_mask)

    # Convert the blue background to grayscale
    gray_frame = cv2.cvtColor(blue_background, cv2.COLOR_BGR2GRAY)

    # Apply the LoG operator to detect blobs on the blue background
    ksize = 1  # Adjust the kernel size based on the size of blobs you want to detect
    sigma = 2.5  # Adjust the sigma value based on the spread of blob intensity
    blobs = cv2.Laplacian(gray_frame, cv2.CV_64F, ksize=ksize)
    blobs = np.uint8(np.absolute(blobs))

    # Threshold the blob image to filter out low-intensity regions
    threshold = 20  # Adjust the threshold value as per your requirements
    _, binary_blobs = cv2.threshold(blobs, threshold, 255, cv2.THRESH_BINARY)

#     # Find contours in the binary blob image
#     contours, _ = cv2.findContours(binary_blobs, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

#     # Draw bounding boxes around the detected blobs on the original frame
#     for contour in contours:
#         x, y, w, h = cv2.boundingRect(contour)
#         cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

#     return frame
    contours, _ = cv2.findContours(binary_blobs, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Find the largest contour (blob) based on its area
    largest_contour = max(contours, key=cv2.contourArea)

    # Draw bounding box around the largest blob on the original frame
    if len(largest_contour) > 0:
        x, y, w, h = cv2.boundingRect(largest_contour)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    return frame

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Perform blob detection on the blue background and draw boundary boxes
    processed_frame = blob_detection_blue_background(frame)

    # Display the processed frame
    cv2.imshow('Processed Frame', processed_frame)

    # Exit the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
